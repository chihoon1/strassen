```python
p = 6912
```

```python
k_arr = np.arange(3,10)
t_arr_seq = np.array([0.0001, 0.0005, 0.0034, 0.0238, 0.1650, 1.1563, 8.0348,])  # t_1
t_arr_gpu = np.array([0.1528, 0.0959, 0.0982, 0.0923, 0.0973, 0.1268, 0.2761,])  # t_p
speed_up_arr = t_arr_seq / t_arr_gpu  # speed up
e_p = speed_up_arr / p  # efficiency
```

```python
print("Matrix has number of elements equal to 2^(n) * 2^(n)")
plt.plot(k_arr, t_arr_seq, label="CPU Sequential Execution")
plt.plot(k_arr, t_arr_gpu, label=f"GPU Parallel Execution; p={p}")
plt.legend()
plt.xlabel("k")
plt.ylabel("Execution Time")
plt.title("Execution Time Comparison")
plt.show()
```

```python
print("Matrix has number of elements equal to 2^(n) * 2^(n)")
fig = plt.figure(figsize=(15,5))
for i, (name, metric_arr) in enumerate(zip(["Speedup", "Efficiency"],[speed_up_arr, e_p])):
    plt.subplot(1, 2, i+1)
    plt.plot(k_arr, metric_arr)
    plt.xlabel("k")
    plt.ylabel(name)
    plt.title(f"{name} of Strassen Recursive Algorithm; p={p}")
```

```python
k_prime_arr = np.arange(1,10)
# t_1
t_arr_seq = np.array([0.1750, 0.1247, 0.1170, 0.1278, 0.1619, 0.2404, 0.4899, 1.5348, 8.0208,])
# t_p
t_arr_gpu = np.array([0.1014, 0.0942, 0.0954, 0.0965, 0.1069, 0.1062, 0.1140, 0.1384, 0.2662,])
speed_up_arr = t_arr_seq / t_arr_gpu  # speed up
e_p = speed_up_arr / p  # efficiency
```

```python
print("Matrix has number of elements equal to 2^(n) * 2^(n)")
print("k' is the max depth level in recursion")
plt.plot(k_prime_arr, t_arr_seq, label="CPU Sequential Execution")
plt.plot(k_prime_arr, t_arr_gpu, label=f"GPU Parallel Execution; p={p}")
plt.legend()
plt.xlabel("k' max depth level")
plt.ylabel("Execution Time")
plt.title("Execution Time Comparison")
plt.show()
```

```python
print("Matrix has number of elements equal to 2^(n) * 2^(n)")
print("k' is the max depth level in recursion")
fig = plt.figure(figsize=(15,5))
for i, (name, metric_arr) in enumerate(zip(["Speedup", "Efficiency"],[speed_up_arr, e_p])):
    plt.subplot(1, 2, i+1)
    plt.plot(k_prime_arr, metric_arr)
    plt.xlabel("k' max depth level")
    plt.ylabel(name)
    plt.title(f"{name} of Strassen Recursive Algorithm; p={p}")
```
